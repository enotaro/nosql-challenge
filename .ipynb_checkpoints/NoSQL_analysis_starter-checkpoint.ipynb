### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20

# Create a query that searches for hygiene score of 20
query = {'scores.Hygiene': 20}

# Have the results only show the Business Name instead of all fields
fields = {"BusinessName": 1}

# Search for the query with the specified fields
hygiene20 = db.establishments.find(query, fields)

# Print the results
for result in hygiene20:
    pprint(result)
```

```python
# Use count_documents to display the number of documents in the result
number_hygiene20 = establishments.count_documents({'scores.Hygiene': 20})
print(number_hygiene20)
```

```python
# Display the first document in the results using pprint

# Limit to just one document
limit = 1

# Find the result of the query with limit of 1
result = list(establishments.find(query).limit(limit))

# Print the result
pprint(result)
```

```python
# Convert the result to a Pandas DataFrame

# Import pandas
import pandas as pd

# Find the query
all_results = db.establishments.find(query)

# Put the results of the query into a dataframe
results_of_hygiene20_df = pd.DataFrame(all_results)
```

```python
# Display the number of rows in the DataFrame
print(len(results_of_hygiene20_df))
```

```python
# Display the first 10 rows of the DataFrame
results_of_hygiene20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

# Create a query that searches for the word London in LocalAuthorityName
# and has a rating value of greater than or equal to 4
query = {"LocalAuthorityName": {'$regex': 'London'},
         "RatingValue": {'$gte': 4}}

# Have the results only show the Business Name field
fields = {"BusinessName": 1}

# Find the query with the specified field
london_rating_4up = db.establishments.find(query, fields)

# Print the result
for result in london_rating_4up:
    pprint(result)
```

```python
# Use count_documents to display the number of documents in the result
number_london_rating_4up = establishments.count_documents({"LocalAuthorityName":  {'$regex': 'London'}, "RatingValue": {'$gte': 4}})
print(number_london_rating_4up)
```

```python
# Display the first document in the results using pprint

# Limit to only 1 result
limit = 1

# Find the result of the query with only 1 result
result = list(establishments.find(query).limit(limit))

# Print the result
pprint(result)
```

```python
# Convert the result to a Pandas DataFrame

# Find the query
all_results = db.establishments.find(query)

# Put results into a dataframe
results_of_london_rating_4up_df = pd.DataFrame(all_results)
```

```python
# Display the number of rows in the DataFrame
results_of_london_rating_4up_df.shape[0]
```

```python
# Display the first 10 rows of the DataFrame
results_of_london_rating_4up_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01

# Find latitude and longitude of Penang Flavours from the information we input in the setup starter notebook
latitude = 0.08384000
longitude = 51.49014200

# Create a query for rating value of 5,
# latitude between +- 0.01 of Penang Flavours' latitude,
# and longitude between +- 0.01 of Penang Flavours' longitude.
query = {'RatingValue': 5,
         'geocode.latitude': {'$in': [0.07384000, 0.09384000]},
         'geocode.longitude': {'$in': [51.48014200, 51.50014200]}}

# Have the results show just the Business Name field
fields = {"BusinessName": 1}

# Sort by the hygiene scores in ascending order
sort = [('scores.Hygiene', 1)]

# Limit to 5 results
limit = 5

# Find the query with the specified fields, sort, and limit
question3 = db.establishments.find(query, fields).sort(sort).limit(limit)

# Print the results
for result in question3:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
all_results = db.establishments.find(query, fields).sort(sort).limit(limit)
results_df = pd.DataFrame(all_results)
```

```python
# Print the first 10 rows of the dataframe
results_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 

# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1}}
               
# Create the pipeline
pipeline = [match_query, group_query, sort_values]

# Create a list of the results
results = list(establishments.aggregate(pipeline))
```

```python
# Print the number of documents in the result
print(len(results))
```

```python
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
question4_df = pd.DataFrame(results)
```

```python
# Display the number of rows in the DataFrame
print(len(question4_df))
```

```python
# Display the first 10 rows of the DataFrame
question4_df.head(10)
```
